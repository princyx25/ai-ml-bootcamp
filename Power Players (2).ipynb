6 Steps of Web Scraping

1. Send a Request to the Website
   The scraper sends an HTTP request to the target website to access its webpage content.

2. Retrieve the HTML Content
   The website responds by sending its HTML source code, which contains the webpage data.

3. Parse the HTML
   The HTML content is analyzed and converted into a structured format that can be easily searched and processed.

4. Locate the Required Data
   Specific HTML elements such as tags, classes, or IDs are identified to find the desired information.

5. Extract and Clean the Data
   The required data is extracted from the webpage and cleaned by removing unnecessary characters, spaces, or formatting.

6. Store the Data
   The extracted data is saved in a suitable format such as CSV, Excel, JSON, or a database for future use.

Flow:
Request Website → Get HTML → Parse HTML → Locate Data → Extract Data → Store Data.

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.flipkart.com/search?q=shoes"

headers = {
    "User-Agent": "Mozilla/5.0"
}

response = requests.get(url, headers=headers)

soup = BeautifulSoup(response.text, "html.parser")

print("Status Code:", response.status_code)
print("Page Title:", soup.title.text)
```

```python
print(soup.prettify()[:5000])
```

```python
products = soup.find_all('div')

print("Total divs:", len(products))
```

```python
for div in products:
    text = div.get_text(" ", strip=True)

    if "₹" in text:
        print(text[:300])
        print("-" * 50)
        break
```

```python
for div in products:
    text = div.get_text(" ", strip=True)

    if "₹" in text and "off" in text and ("Running Shoes" in text or "Sneakers" in text or "Shoes" in text):
        print(text[:500])
        print("-" * 80)
```

```python
shoes_data = []

for div in products:
    text = div.get_text(" ", strip=True)

    if "₹" in text and "off" in text:
        shoes_data.append(text)

for item in shoes_data[:10]:
    print(item)
    print("-" * 80)
```

```python
for div in products:
    text = div.get_text(" ", strip=True)

    if "₹" in text and len(text) > 100:
        print(text[:1000])
        print("=" * 100)
```

```python
print(soup.title.text)

for tag in soup.find_all(['div','a'])[:50]:
    if tag.get('class'):
        print(tag.name, tag.get('class'))
```

```python
for tag in soup.find_all("a"):
    text = tag.get_text(" ", strip=True)

    if text:
        print("TEXT:", text[:100])
        print("CLASS:", tag.get("class"))
        print("-" * 50)
```

```python
names = soup.find_all("a", class_="atJtCj")
prices = soup.find_all("a", class_="k7mUBC")

for i in range(min(len(names), len(prices))):
    print("Product Name:", names[i].get_text(strip=True))
    print("Price Details:", prices[i].get_text(" ", strip=True))
    print("-" * 80)
```

```python
import pandas as pd

data = []

names = soup.find_all("a", class_="atJtCj")
prices = soup.find_all("a", class_="k7mUBC")

for i in range(min(len(names), len(prices))):
    data.append({
        "Product Name": names[i].get_text(strip=True),
        "Price Details": prices[i].get_text(" ", strip=True)
    })

df = pd.DataFrame(data)
df
```
